```python
# imports
import pandas as pd
import numpy as np
import requests
import bs4
import time as t
import random as ran
```

```python
# lists
names = []
review_counts = []
rating = []
price_range = []
price_range_count = []
phone_number = []
address = []
```

```python
# scraper set-up
pages = np.arange(0, 100, 10)
url = 'https://www.yelp.com/search?find_desc=Restaurants&find_loc=London%2C%20United%20Kingdom&ns=1&start='
container_class = {'class' : 'container__09f24__21w3G'}
header = {
'Accept-Language': 'en-US,en;q=0.9,es;q=0.8',
'Upgrade-Insecure-Requests': '1',
'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6)Mozilla/5.0 (Macintosh; Intel Mac OS X 10_12_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'
}
```

```python
# scrape pages
for page in pages:
    page = requests.get(url + str(page), headers = header)
    scraper = bs4.BeautifulSoup(page.text)
    div = scraper.find_all('div', container_class)
    t.sleep(ran.randint(2,10))
    
    # scrape within div container    
    for container in div:
        
        # find restaraunt names
        names_class = {'class' : 'link-size--inherit__09f24__2Uj95'}
        name = container.find('a', names_class).text
        names.append(name)
        
        # find rating
        rating_class = {'class' : 'i-stars__09f24__1T6rz'}
        stars = container.find('div', rating_class)['aria-label']
        rating.append(stars)
        
        # find review counts
        revCount_class = {'class' : 'reviewCount__09f24__EUXPN'}
        revCount = container.find('span', revCount_class).text
        review_counts.append(revCount)
        
        # find price range
        try: 
            price_class = {'class' : 'priceRange__09f24__2O6le'}
            price = container.find('span', price_class).text
            price_range.append(price)
        except:
            price_range.append(None)

        # find contact phone number
        phone_class = {'class' :  'text__09f24__2tZKC'}
        phone = container.find('p', phone_class).text
        phone_number.append(phone)
        
        # find address
        addy_class = {'class' : 'raw__09f24__3Obuy'}
        addy = container.find('span', addy_class).text
        address.append(addy)
  
```

```python
# star rating formatting
stars = []
for string in rating:
    string = string.replace('star rating', 'stars')
    stars.append(string)
```

```python
# contact number formatting
contact_number = []
for number in phone_number:
    try:
        number = number.replace(' ','')
        number = int(number)
        number = '0' + str(number)
        contact_number.append(number)
    except:
        contact_number.append(None)
```

```python
# price range count
price_range_count =[]
for p in price_range:
    if p is None:
        price_count = None
        price_range_count.append(price_count)
    else:
        price_count = p.count('\xA3')
        price_range_count.append(price_count)
```

```python
# create dictionary
yelp_dict = {
    'Name' : names,
    'Rating' : stars,
    'Number of Reviews' : review_counts,
    'Price range' : price_range,
    'Price range count' : price_range_count,
    'Contact number' : contact_number,
    'Street Address' : address
}
```

```python
# create dataframe
df = pd.DataFrame(yelp_dict)
pd.set_option("max_rows", None)
df
```
